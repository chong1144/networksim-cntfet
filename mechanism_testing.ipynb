```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as const
```

```python
# Schottky current due to thermal. from svennson
#thermal voltage, defined as Kb*T/e
vt=300*const.Boltzmann/const.elementary_charge
def schottky_current(vd,vg,sbe=0.45,n=1.2):
    # 1<n<2 ideality factor. 1 for ideal
    # Isat =AA*T^2 exp(-SBe/VT)
    A=(1e-10)**2
    Aconst=4*np.pi*const.electron_mass*const.elementary_charge*const.Boltzmann**2/const.h**3
    Isat = A*Aconst*300**2*np.exp((vg-sbe)/vt)
    
    return Isat*np.exp(vd/(n*vt))*(1-np.exp(-vd/vt))
    
```

```python
x=np.linspace(-0.1,0.1,100)
y=[abs(schottky_current(i,1)) for i in x]
plt.semilogy(x,y)
```
