```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import viewnet 
import networkx as nx
```

```python
collection=viewnet.Netviewer(n=300,scaling=5,notes='presentation',l=0.66)
```

```python
collection.cnet.set_global_gate(0)
collection.cnet.gate_areas=[]
collection.cnet.update()
collection.show_system(save="talk_example_on")
sum(collection.cnet.source_currents)
```

```python
collection.cnet.set_global_gate(0)
collection.cnet.gate_areas=[]
collection.cnet.set_local_gate([0.3,1,0.4,1.2], 10)
collection.cnet.update()
collection.show_system(save="talk_example_partialoff")
sum(collection.cnet.source_currents)
```

```python
collection.cnet.set_global_gate(0)
collection.cnet.gate_areas=[]
collection.cnet.set_global_gate(10)
collection.cnet.update()
collection.show_system(save="talk_exampletotaloff")
sum(collection.cnet.source_currents)
```

```python
nclust=len(collection.sticks.cluster.drop_duplicates())
maxclust=len(max(nx.connected_components(collection.graph)))
print(nclust,maxclust)
```
